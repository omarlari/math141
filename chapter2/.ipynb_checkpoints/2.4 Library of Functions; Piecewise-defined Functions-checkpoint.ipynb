#### 25.) graph the function $f(x)=\sqrt[3]{x}$

```python
import matplotlib.pyplot as plt
import numpy as np

def f1(x):
    answer = x**(1/float(3))
    return answer

def drawGraph(x,y):
    plt.plot(x, y)
    plt.scatter(1, f1(1),c='red')
    plt.scatter(8, f1(8),c='red')
    plt.scatter(0, f1(0),c='red')
    plt.text(1, f1(1), '(1,1)')
    plt.text(0, f1(0), '(0,0)')
    plt.text(8, f1(8), '(2,8)',verticalalignment='top')
    plt.grid(True)
    plt.show()

def createArrays():
    intervals = np.arange(0,10,.001)
    a = []
    b = []
    for t in intervals:
        a.append(t)
        b.append(f1(t))
    x = np.array(a)
    y = np.array(b)
    return x, y

drawGraph(createArrays()[0], createArrays()[1])

print(f1(1))
```
