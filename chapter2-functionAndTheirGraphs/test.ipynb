```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#plt.rcdefaults()
x = np.arange(-10, 10, .1)

def f1(x):
    answer = x**2
    return answer

# a < 0
def f2(x):
    answer = 1/2*x**2
    return answer

# a > 0
def f3(x):
    answer = 2*x**2
    return answer

plt.figure(figsize=(10,10))


plt.plot(x, f1(x), c='red')
plt.plot(x, f2(x), c='blue')
plt.plot(x, f3(x), c='green')

# optics
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.xticks(np.arange(-10,10,step=1))
plt.yticks(np.arange(-10,10))
plt.grid(True)


plt.show()
```
